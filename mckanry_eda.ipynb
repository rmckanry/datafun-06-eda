```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import csv
```

```python
#Read and display dataset CSV file, showing the first 10 rows
df = pd.read_csv("Electric_Vehicle_Population_Data.csv")
print(df.head(10))
print(df.shape)
print(df.dtypes)
print(df.info())

print(df.describe())
```

```python
# Create a chart showing the total number of EVs owned between 1999 and 2025
plt.figure(figsize=(12,6))
sns.countplot(x = 'Make', data = df)

# Readability
plt.xticks(rotation=45, ha='right')

# Title and Labels
plt.title('Distribution of EVs by Maker 1999-2025')
plt.ylabel('Total EVs')
plt.xlabel("EV Manufacturer")

# Show Plot
plt.show()
```

```python
# Display scatter plot showing the year an auto maker started building EVs
#Set size
plt.figure(figsize=(8,15))

#Set titles and labels
plt.title('EVs Manufacturer by Years Produced 1999-2025')
plt.ylabel("EV Manufacturer")
plt.xlabel("Production Year")

#Create plot and show
sns.scatterplot(data=df, x="Model Year", y="Make", hue="Make")
plt.show()
```
