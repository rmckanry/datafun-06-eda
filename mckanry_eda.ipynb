```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import csv
```

```python
#df = sns.load_dataset("Electric_Vehicle_Population_Data.csv")
#df = sns.load_dataset('iris')
df = pd.read_csv("Electric_Vehicle_Population_Data.csv")
print(df.head(10))
print(df.shape)
print(df.dtypes)
print(df.info())

print(df.describe())
```

```python
# Inspect histogram by numerical column
#df['Model Year'].hist()

# Inspect histograms for all numerical columns
#df.hist()

# Show all plots
#plt.show()
'''
# Inspect value counts by categorical column
df['State'].value_counts()

# Inspect value counts for all categorical columns
for col in df.select_dtypes(include=['object', 'category']).columns:
    # Display count plot
    sns.countplot(x=col, data=df)
    plt.title(f'Distribution of {col}')
    plt.show()

# Show all plots
plt.show()
'''
#y = df['Make'].array
#print("Y=",y)

# Create a box plot showing the distribution to remind us of the distribution
plt.figure(figsize=(12,6))
sns.countplot(x = 'Make', data = df)

# Readability
plt.xticks(rotation=45, ha='right')

# Title and Labels
plt.title('Distribution of EVs by Maker 1999-2025')
plt.ylabel('Total EVs')
plt.xlabel("EV Manufacturer")

# Show Plot
plt.tight_layout()
plt.show()
```

```python
# Display scatter plot for Goal 3
plt.figure(figsize=(8,15))
plt.title('EVs Manufacturer by Years Produced 1999-2025')
plt.ylabel("EV Manufacturer")
plt.xlabel("Production Year")
sns.scatterplot(data=df, x="Model Year", y="Make", hue="Make")
plt.show()
```
